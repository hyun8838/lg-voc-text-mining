```python
import os
import pandas as pd

# 작업 경로 지정
data_path = "C:/Users/legen/Desktop/Lab Project/LG text mining/data/"

# 카테고리와 타입 리스트
categories = ["의류관리기", "무선스틱청소기", "로봇청소기"]
types = ["cafe1", "cafe2", "cafe3", "blog", "youtube"]
```

```python
# 파일을 읽고 통합하는 함수
def load_and_combine(category, types, data_path):
    file_list = []  # 파일 리스트 초기화
    
    for t in types:
        # 파일 경로 생성
        file_name = f"{category}_{t}.csv"
        file_path = os.path.join(data_path, file_name)
        
        # 파일이 존재하면 읽기
        if os.path.exists(file_path):
            df = pd.read_csv(file_path)
            file_list.append(df)
        else:
            print(f"File not found: {file_path}")
    
    # 리스트에 저장된 데이터프레임들을 결합
    combined_data = pd.concat(file_list, ignore_index=True)
    return combined_data
```

```python
# 의류관리기, 무선청소기, 로봇청소기 각각에 대해 통합
의류관리기_pool = load_and_combine("의류관리기", types, data_path)
무선청소기_pool = load_and_combine("무선스틱청소기", types, data_path)
로봇청소기_pool = load_and_combine("로봇청소기", types, data_path)
```

```python
# 통합된 데이터를 csv 파일로 저장
의류관리기_pool.to_csv(os.path.join(data_path, "의류관리기_pool.csv"), index=False, encoding='utf-8-sig')
무선청소기_pool.to_csv(os.path.join(data_path, "무선스틱청소기_pool.csv"), index=False, encoding='utf-8-sig')
로봇청소기_pool.to_csv(os.path.join(data_path, "로봇청소기_pool.csv"), index=False, encoding='utf-8-sig')


print("CSV 파일이 성공적으로 저장되었습니다.")
```
