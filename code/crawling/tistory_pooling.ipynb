```python
import os
import pandas as pd

# 데이터 불러오고 병합 함수
def load_and_process_data(folder_path):
    # 경로 내 파일 목록 가져오기
    files = os.listdir(folder_path)
    
    # 'tistory_종류_script' 패턴 파일 필터링
    tistory_files = [file for file in files if file.startswith('tistory_') and file.endswith('.csv')]
    
    # 데이터를 담을 딕셔너리
    data_vars = {}
    
    # 파일별 데이터프레임 생성
    for file in tistory_files:
        # 파일 경로
        file_path = os.path.join(folder_path, file)
        
        # 파일명에서 변수명 추출 (확장자 제거)
        variable_name = file.split('.')[0]
        
        # 데이터 불러오기
        df = pd.read_csv(file_path)
        
        # 열 이름 변경 및 platform 열 추가
        df.rename(columns={'url': 'link'}, inplace=True)
        df['platform'] = 'tistory'
        
        # 변수 저장
        data_vars[variable_name] = df
    
    # 데이터 병합 작업
    merged_data = {}
    merge_groups = [
        'styler', 'airdresser', 'lg_vacuum', 'samsung_vacuum', 'dyson_vacuum', 'robot'
    ]
    
    for group in merge_groups:
        # 그룹에 해당하는 데이터프레임 병합
        dfs_to_merge = [
            data_vars[key]
            for key in data_vars if key.startswith(f"tistory_{group}_script")
        ]
        if dfs_to_merge:
            merged_data[f"tistory_{group}_script"] = pd.concat(dfs_to_merge, ignore_index=True)
    
    return data_vars, merged_data
```

```python

# 실행
folder_path = "C:/Users/legen/Desktop/Lab Project/LG text mining/python/tistory"
data_vars, merged_data = load_and_process_data(folder_path)

# 개별 데이터 확인
for key, df in data_vars.items():
    print(f"Dataset: {key}")
    print(df.head(), "\n")

# 병합된 데이터 확인
for key, df in merged_data.items():
    print(f"Merged Dataset: {key}")
    print(df.head(), "\n")
```

```python
# 결과 저장 (선택 사항)
for key, df in merged_data.items():
    output_path = os.path.join(folder_path, f"{key}_merged.csv")
    df.to_csv(output_path, index=False, encoding='utf-8-sig')
    print(f"{key} 병합 데이터가 {output_path}에 저장되었습니다.")
```

## 데이터통합 ##

```python
import os
import pandas as pd

# 데이터 불러오고 처리 함수
def load_and_process_data(folder_path):
    # 경로 내 파일 목록 가져오기
    files = os.listdir(folder_path)
    
    # 파일별 데이터 저장 딕셔너리
    datasets = {}
    
    # 모든 파일 불러오기
    for file in files:
        if file.endswith('.csv'):
            file_path = os.path.join(folder_path, file)
            variable_name = file.split('.')[0]  # 확장자 제거 후 변수명 생성
            datasets[variable_name] = pd.read_csv(file_path)

    # tistory_로 시작하고 _script로 끝나는 데이터 중복 제거 및 열 순서 변경
    for key in [k for k in datasets.keys() if k.startswith("tistory_") and k.endswith("_script")]:
        df = datasets[key]
        
        # link 열이 있는지 확인
        if "link" in df.columns:
            # 중복 제거
            df = df.drop_duplicates(subset="link").reset_index(drop=True)
            # 열 순서 변경
            desired_columns = ["title", "content", "date", "link", "platform"]
            # 모든 열이 존재하는 경우에만 재배치
            if all(col in df.columns for col in desired_columns):
                df = df[desired_columns]
            else:
                print(f"Warning: {key} 데이터에 필요한 열이 없습니다. (필요한 열: {desired_columns})")
        else:
            print(f"Warning: {key} 데이터에 'link' 열이 없습니다. 해당 데이터를 건너뜁니다.")
            continue
        
        # 변경된 데이터 저장
        datasets[key] = df

    # _pool로 끝나는 데이터에 tistory 데이터를 통합
    # 의류관리기_pool
    if "의류관리기_pool" in datasets:
        datasets["의류관리기_pool"] = pd.concat(
            [datasets["의류관리기_pool"],
             datasets.get("tistory_styler_script", pd.DataFrame()),
             datasets.get("tistory_airdresser_script", pd.DataFrame())],
            ignore_index=True
        )

    # 무선스틱청소기_pool
    if "무선스틱청소기_pool" in datasets:
        datasets["무선스틱청소기_pool"] = pd.concat(
            [datasets["무선스틱청소기_pool"],
             datasets.get("tistory_lg_vacuum_script", pd.DataFrame()),
             datasets.get("tistory_samsung_vacuum_script", pd.DataFrame()),
             datasets.get("tistory_dyson_vacuum_script", pd.DataFrame())],
            ignore_index=True
        )

    # 로봇청소기_pool
    if "로봇청소기_pool" in datasets:
        datasets["로봇청소기_pool"] = pd.concat(
            [datasets["로봇청소기_pool"],
             datasets.get("tistory_robot_script", pd.DataFrame())],
            ignore_index=True
        )
    
    return datasets
```

```python
# 실행
folder_path = "C:/Users/legen/Desktop/Lab Project/LG text mining/python/data"
processed_datasets = load_and_process_data(folder_path)
```

```python
# 결과 확인 및 저장 (선택사항)
output_folder = os.path.join(folder_path, "processed")
os.makedirs(output_folder, exist_ok=True)

for key, df in processed_datasets.items():
    print(f"Dataset: {key}")
    print(df.head(), "\n")
    # 저장
    df.to_csv(os.path.join(output_folder, f"{key}.csv"), index=False, encoding='utf-8-sig')
    print(f"{key} 데이터 저장 완료!")
```
